```python
num_loops = 100000
num_steps = 5
num_states = 4
num_actions = 5
prev_policy = np.ones([num_states * num_steps], dtype=int)  # make different from policy
policy = np.zeros([num_states * num_steps], dtype=int)
sum_rewards = np.zeros([num_actions, num_states * num_steps])
loop_counts = np.zeros([num_actions, num_states * num_steps])
epsilon = 0.05
```

```python
import random
print(policy)
while not np.array_equal(prev_policy, policy):
    for i in range(num_loops):
        start_state = random.randint(0,num_states - 1)
        state = np.zeros([num_states * num_steps])
        state[start_state] = 1
        total_reward = 0
        state_list = np.zeros([num_steps], dtype=int)
        action_list = np.zeros([num_steps], dtype=int)
        for step_idx in range(num_steps):
            #print(state)
            state_idx = np.argmax(state)
            if random.uniform(0, 1) > epsilon:
                action = policy[state_idx]
            else:
                action = random.randint(0, num_actions - 1)
            state_list[step_idx] = state_idx
            action_list[step_idx] = action
            #print(action)
            #print(state_idx)
            #print(step_idx)
            total_reward += r[action][state_idx]
            state = np.dot(P[action].T, state)
        for idx in range(len(state_list)):
            sum_rewards[action_list[idx], state_list[idx]] += total_reward
            loop_counts[action_list[idx], state_list[idx]] += 1
    avg_rewards = sum_rewards / loop_counts
    prev_policy = policy
    policy = avg_rewards.argmax(axis=0)
    print(policy)
avg_rewards = sum_rewards / loop_counts
print(avg_rewards.max(axis=0))
print(avg_rewards)
```
